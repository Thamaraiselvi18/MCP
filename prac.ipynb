```python
import gspread
from google.oauth2.service_account import Credentials

# ---------------------------
# AUTHENTICATION
# ---------------------------
scopes = [
    "https://www.googleapis.com/auth/spreadsheets",
    "https://www.googleapis.com/auth/drive"
]

creds = Credentials.from_service_account_file(
    "credentials.json",
    scopes=scopes
)

client = gspread.authorize(creds)

# ---------------------------
# OPEN SHEET
# ---------------------------
sheet_id = "1Yvvd-P8nqCVEtVnEuK8u3eDMiVKSaVE9fYMMGTf0ezk"
workbook = client.open_by_key(sheet_id)

# ---------------------------
# DATA TO INSERT
# ---------------------------
values = [
    ["Name", "Age", "City"],
    ["Alice", 30, "New York"],
    ["Bob", 25, "Los Angeles"],
    ["Charlie", 35, "Chicago"]
]

# ---------------------------
# GET / CREATE WORKSHEET
# ---------------------------
worksheet_list = [ws.title for ws in workbook.worksheets()]
new_worksheet_name = "Values"

if new_worksheet_name in worksheet_list:
    sheet = workbook.worksheet(new_worksheet_name)
else:
    sheet = workbook.add_worksheet(
        title=new_worksheet_name,
        rows="10",
        cols="10"
    )

# ---------------------------
# UPDATE DATA
# ---------------------------
sheet.clear()

sheet.update(
    values=values,
    range_name=f"A1:C{len(values)}"
)

print("✅ Data updated successfully!")

    
```

```python
import gspread
from google.oauth2.service_account import Credentials

# ---------------------------
# AUTH
# ---------------------------
scopes = [
    "https://www.googleapis.com/auth/spreadsheets",
    "https://www.googleapis.com/auth/drive"
]

creds = Credentials.from_service_account_file(
    "credentials.json",
    scopes=scopes
)

client = gspread.authorize(creds)

# ---------------------------
# OPEN WORKBOOK
# ---------------------------
sheet_id = "1Yvvd-P8nqCVEtVnEuK8u3eDMiVKSaVE9fYMMGTf0ezk"
workbook = client.open_by_key(sheet_id)

# ---------------------------
# USE EXISTING SHEET (Sheet1)
# ---------------------------
sheet = workbook.sheet1   # ✅ THIS IS THE KEY LINE

# ---------------------------
# DATA
# ---------------------------
values = [
    ["Name", "Age", "City"],
    ["Alice", 30, "New York"],
    ["Bob", 25, "Los Angeles"],
    ["Charlie", 35, "Chicago"]
]

# ---------------------------
# UPDATE DATA
# ---------------------------
sheet.clear()

sheet.update(
    values=values,
    range_name=f"A1:C{len(values)}"
)

print("✅ Data written into existing Sheet1")
```
